```python
import mlflow
import pandas as pd
import numpy as np

import argparse
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Conv2D
from tensorflow.keras import Model
from tensorflow.keras.datasets.mnist import load_data
from ray import tune
import os
```

```python
DATA_PATH = '/Users/camilovelasquez/Desktop/Documents/Datasets/WISDM-Smartphones/wisdm-dataset/raw'
```

```python
ids = [1600, 1601]
devices = ['phone', 'watch']
sensors = ['accel', 'gyro']
```

```python
current_id = ids[0]
current_device = devices[0]
current_sensor = sensors[0]
```

```python
i = 0
for current_id in ids:
    for current_device in devices:
        for current_sensor in sensors:
            file_path = os.path.join(DATA_PATH, current_device, current_sensor, 
                                     'data_{}_{}_{}.txt'.format(current_id, current_sensor, current_device))
            if i==0:
                table = pd.read_csv(file_path, delimiter=',', names=['ID', 'Activity Label', 'Timestamp', 'x', 'y', 'z'])
            else:
                aux = pd.read_csv(file_path, delimiter=',', names=['ID', 'Activity Label', 'Timestamp', 'x', 'y', 'z'])
                table = pd.concat([table, aux], axis=0)
            i+=1
```
